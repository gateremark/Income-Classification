```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("income_evaluation.csv")
data.head()
```

```python
data = data.rename(columns={" fnlwgt": "final_weight"})
data.head()
```

```python
data[" income"].unique()
```

```python
data.info()
```

```python
# splitting into x and y
x = data.drop(" income", axis = 1)
y = data[" income"]
```

```python
# converting features into numerical using dummies
dummies = pd.get_dummies(x)
dummies
```

```python
#splitting into training and testing set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dummies, y, test_size = 0.2)
```

```python
# selecting a model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()

model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
# using the metrics

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)#, labels =[" <=50K", " >50K"])
```
